# Data Binning

```python
df[['horsepower']].head()
```

```python
bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
bins
```

```python
categories_names = ['Low', 'Medium', 'High']

df['horsepower-binned']=pd.cut(df['horsepower'], bins, labels = categories_names, include_lowest=True)
df[['horsepower', 'horsepower-binned']].head(30)
```

```python
bins1 = np.linspace(min(df['price']),max(df['price']),4)
bins1
```

```python
df['price-binned']=pd.cut(df['price'], bins1, labels = categories_names, include_lowest=True)
df[['price', 'price-binned']].head(30)
```
